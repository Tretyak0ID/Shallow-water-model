```python
%load_ext autoreload
%autoreload 2
import numpy as np
from matplotlib import pyplot as plt
#deriv module
#diff_c2_1, diff_c2_2, diff_c2_3, diff_c2_4 - 2nd order central difference functions
#diff_c4_1, diff_c4_2, diff_c4_3. diff_c2_4 - 4th order central difference functions
import deriv #my library for differentiation
```

```python
def diff_c2_n(f, h):
    diff_f = np.zeros(f.size)
    for i in range(1,f.size-1):
        diff_f[i] = (f[i+1]-f[i-1])/ 2 / h
    diff_f[-1] = (f[1]-f[-2])/ 2 / h
    diff_f[0] = diff_f[-1]
    return diff_f 
```

```python
# test derivative calculation accuracy, return linf, l2 error norms and analytic and numerical derivative
# input: diff - numerical differentiation function to be tested,
#               diff interface diff(fun_np_array,grid_step)
#        test_fun - derivative of this fucnction will be calculated
#        test_fun_deriv - function to calculate derivation of test_fun exactly
#        n - number of points in grid [a,b), NB: b point is excluded
def test_deriv(diff,test_fun,test_fun_deriv,n,a=-np.pi,b=np.pi):
    x        = np.linspace(a,b,n,endpoint=False)
    h        = (b-a)/n
    f        = test_fun(x)
    f1_exact = test_fun_deriv(x)
    f1       = diff(f,h)
    
    e = f1-f1_exact
    linf = np.max(np.abs(e))
    l2   = np.sqrt(np.sum(e**2)/n)
    return (linf, l2, f1, f1_exact)
```

```python
def plot_convergence(n_list,diff,assumed_order=2,test_fun=np.sin,test_fun_deriv= np.cos):
    l2_array   = np.empty(len(n_list))
    linf_array = np.empty(len(n_list))
    
    for i,n in enumerate(n_list):
        linf, l2, f1, f1_exact = test_deriv(diff,test_fun,test_fun_deriv,n)
        l2_array[i] = l2
        linf_array[i] = linf

    #assumed convergence line
    conv = 0.5*(l2_array[0]+linf_array[0])*(n_list[0]/np.array(n_list))**(assumed_order)
    
    print(f1-f1_exact)

    fig = plt.figure()
    plt.loglog(np.array(n_list),l2_array,label="l2")
    plt.loglog(np.array(n_list),linf_array,label="linf")
    plt.loglog(np.array(n_list),conv,label="conv")
    plt.legend()
    plt.show()
    
    fig = plt.figure()
    plt.plot(f1)
    plt.legend()
    plt.show()
    
    fig = plt.figure()
    plt.plot(f1-f1_exact)
    plt.legend()
    plt.show()
```

### Central 2

```python
#2_0    
plot_convergence([10,20,40], deriv.diff_c2_0, assumed_order=2, test_fun=np.sin, test_fun_deriv=np.cos)
```
